```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
import statsmodels.api as sm
from scipy.stats import pearsonr

# --- data ---
data = pd.DataFrame({
    "clade": [
        "Anthropoids", "Strepsirrhines",
        "Perissodactyls", "Odontoceti",
        "Artiodactyls", "Feliformes", "Caniformes", "Proboscidae", "Sciuromorpha", "Hystricognathi", "Hipposideridae"
    ],
    "encephalization_slope": [0.0160, 0.0090, 0.0086, 0.0060, 0.0038, 0.0010, 0.0043, 0.0156, -0.0002, 0.0018, 0.0048],
    "mean_EBL_count": [34, 8, 7, 4.57, 5, 4, 4.2, 24, 1, 4.8, 3]
})

# --- correlation ---
r, p = pearsonr(data["encephalization_slope"], data["mean_EBL_count"])

# --- plot ---
fig, ax = plt.subplots(figsize=(7, 5))
ax.scatter(data["encephalization_slope"], data["mean_EBL_count"], color="tab:grey")

# label each point
for _, row in data.iterrows():
   ax.text(row["encephalization_slope"]*1.02, row["mean_EBL_count"]+0.3,
           row["clade"], fontsize=9)

# regression line
m, b = np.polyfit(data["encephalization_slope"], data["mean_EBL_count"], 1)
x = np.linspace(data["encephalization_slope"].min()*0.9,
                data["encephalization_slope"].max()*1.1, 100)
ax.plot(x, m*x + b, color="gray", linestyle="--")

ax.set_xlabel("Encephalization slope")
ax.set_ylabel("Mean EBL count")
ax.set_title(f" (r = {r:.2f}, p = {p:.3f})")

plt.tight_layout()
plt.show()
```

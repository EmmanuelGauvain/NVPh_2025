
# Calculating new encephalization slopes




## 1) Sciuromorpha


```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

df = pd.read_csv("BrainAllometrySciuromorpha.csv")

# log-transform
df["log_brain"] = np.log10(df["Mean_brain_mass_g"])
df["log_body"] = np.log10(df["Mean_body_mass_g"])

X = sm.add_constant(df[["log_body", "Date"]])
y = df["log_brain"]

model = sm.OLS(y, X).fit()
print(model.summary())
```
